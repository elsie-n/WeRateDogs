# We-Rate-Dogs

The dataset wrangled analyzed and visualized in this project is from a tweet archive of Twitter user @dog_rates(WeRateDogs).WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. 

WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog.WeRateDogs asks people to send photos of their dogs, then tweets selected photos rating and a humorous comment. Dogs are rated on a scale of one to ten, but are invariably given ratings in excess of the maximum, such as "13/10". Popular posts are re-posted on Instagram and Facebook.


### Visualizing Retweet Counts vs Stage

The stages represent the growth of dogs.
The different stages of dogs have different counts of retweets the doggo stage having the most retweet counts as the pupper has the least retweet count.This could be because dogs at the doggo stage are lovable they a a bit bigger than those in the pupper stage having the least retweet count.

### Visualizing retweet count vs favourite count

Retweet count is positively correlated with favourite count,as favourite count increases the number of retweets increase.Retweted tweets tend to appear at the feeds of most account users hence increasing its counts.


### Visualizing Retweet Counts vs Rating Ratio

The ratings of the tweets and retweet count could be said to be positively correlated to each other,the retweet count increases as the ratings increase.
The high rating ratios have high retweet counts unlike the low rated ratios,this could also be correlated to the stages of the dogs since the ratings are based on dog stages too.
